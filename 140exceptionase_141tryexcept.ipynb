```python
intList = range(10)
idx = 10
try:
  print(intList[idx])
except Exception as e:
  exceptionClass = e.__class__.__name__
  exceptionDetail = e.__str__()
  print(f'예외종료:{exceptionClass},예외내용:{exceptionDetail}')


```

```python
intList = range(10)
try:
  idx = int(input('숫자를 입력하세요: '))
  print(intList[idx])
except ValueError:
  print('입력한 내용이 숫자가 아닙니다!')
except IndexError as e:
  print(f'인덱스 오류가 발생: {e}')
```
