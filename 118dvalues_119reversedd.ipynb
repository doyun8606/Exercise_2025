```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
levels = greats.values()
print(list(levels))

greats['조운'] = 97
print(list(levels))
```

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
levels = reversed(greats)
print(list(levels))
```
