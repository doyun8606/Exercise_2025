```python
from os import rename

src = 'sunrise_copied.jpg'
dst = 'sunrise2.jpg'
rename(src,dst)

src = 'sunrise2.jpg'
dst = 'tmp/sunrise2.jpg'
rename(src,dst)
```

```python
from os import listdir
folder = './'
ret = listdir(folder)
print(ret)

from glob import glob
target = '*.txt'
ret = glob(target)
print(ret)

target = '**'
ret = glob(target, recursive=True)
ret.sort()
for r in ret:
  print(r)
```
