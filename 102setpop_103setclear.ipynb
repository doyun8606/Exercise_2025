```python
fruits={'사과','배','오렌지','수박'}
choise = fruits.pop()
print(f'++추출한 요소:{choise}\n++ fruits:{fruits}')

for _ in range(len(fruits)):
  choise = fruits.pop()
  print(f'++추출한 요소:{choise}\n++ fruits:{fruits}')
choise = fruits.pop()
```
