```python
data1 = 'This is data1'
data2 = 'This is data2'
with open('mydata.txt','w') as f:
  f.writelines(data1)
  f.writelines(data2)

with open('mydata.txt','r') as f:
  data = f.read()
print(data)

data1 = 'This is data1\n'
data2 = 'This is data2\n'

data = ['This is data1\n', 'This is data2\n']
with open('mydata.txt','w') as f:
  f.writelines(data)

with open('mydata.txt','r') as f:
  data = f.read()
print(data)
```

```python
bufsize = 1024
with open('sunrise.jpg','rb') as f:
  with open('sunrise_copied.jpg','wb') as h:
    data = f.read(bufsize)
    while data:
      h.write(data)
      data = f.read(bufsize)
```
