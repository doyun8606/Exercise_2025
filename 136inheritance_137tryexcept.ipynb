```python
class Character:
    def __init__(self):          # 생성자에서 캐릭터의 공통 능력치를 1로 초기화
        self.strength = 1        # 힘
        self.offense_power = 1   # 공격력
        self.intelligence = 1    # 지능
        self.mentality = 1       # 정신력
        self.agility = 1         # 민첩성

class Wizard(Character):
    def setBasic(self, name=None, level=1, hp=100):
        self.name = name     # 캐릭터 이름 설정
        self.level = level   # 캐릭터 레벨 설정
        self.hp = hp         # 캐릭터 체력 설정

player1 = Wizard()
player1.setBasic('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')

class Wizard(Character):
    def __init__(self, name=None):
        self.name = name     # 캐릭터 이름
        self.level = 1       # 초기 레벨 1
        self.hp = 100        # 초기 체력 100

'''player1 = Wizard('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')'''

class Wizard(Character):
    def __init__(self, name=None):
        super().__init__()   # 부모 클래스(Character)의 생성자 호출
        self.name = name     # 캐릭터 이름 설정
        self.level = 1       # 초기 레벨 1
        self.hp = 100        # 초기 체력 100

player1 = Wizard('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')

class Parent1:
    def __init__(self):
        print('부모클래스1 생성자')


class Parent2:
    def __init__(self):
        print('부모클래스2 생성자')


class Child(Parent1, Parent2):
    pass


obj = Child()
```
