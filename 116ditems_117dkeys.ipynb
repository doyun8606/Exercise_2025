```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
generals = greats.items()
print(list(generals))

del greats['관우']
print(list(generals))
```

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
general_names = greats.keys()
print(list(general_names))

del greats['관우']
for i, name in enumerate(general_names, start=1):
  print(f'남아 있는 촉나라 장수[{i}]:{name}')
```
