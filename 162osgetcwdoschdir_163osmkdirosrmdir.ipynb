```python
from os import getcwd

print(f'현재 작업 디렉토리: {getcwd()}')

from os import chdir

print(f'현재 작업 디렉토리:{getcwd()}')
chdir('..')
print(f'현재 작업 디렉토리:{getcwd()}')
chdir('c:/mywork/jupyter')
print(f'현재 작업 디렉토리:{getcwd()}')
```

```python
from os import mkdir
folder = 'myfolder'
mkdir(folder)

folder = 'c:/mywork/jupyter/myfolder2'
mkdir(folder)

from os import rmdir

folder = 'myfolder'
folder2 = 'c:/mywork/jupyter/myfolder2'
rmdir(folder)
rmdir(folder2)
```
